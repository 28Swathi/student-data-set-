```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the student data from the CSV file
data = pd.read_csv("student data set.csv")

# Extract relevant columns
names = data['Name']
math_scores = data['Math Score']
english_scores = data['English Score']
science_scores = data['Science Score']

# Create bar charts for scores
plt.figure(figsize=(10, 6))
plt.bar(names, math_scores, color='blue', alpha=0.7, label='Math Score')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Math Scores by Student')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(names, english_scores, color='green', alpha=0.7, label='English Score')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('English Scores by Student')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(names, science_scores, color='red', alpha=0.7, label='Science Score')
plt.xlabel('Student Name')
plt.ylabel('Score')
plt.title('Science Scores by Student')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
